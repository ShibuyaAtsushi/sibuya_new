```python
# 導関数
def divf(x,y):
    return  x * y

# Euler法
def euler(x, y, h, b):
  x_ap = [x]
  y_ap = [y]
  #配列を作りつつ，初期値を入れておく

  while x <= b: #xがbになるまで繰り返し
    
    y =y + (h * divf(x,y))
    x += h
    y_ap.append(y)
    x_ap.append(x)
    
  return x_ap, y_ap

import matplotlib.pyplot as plt
import numpy as np

# グラフ領域の定義
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(1,1,1)

h = 0.01
b = 0.5

# オイラー法のグラフ
x_a , y_a = euler(0.0, 1.0, h, b)
ax.scatter(x_a, y_a, c='purple', label='Euler')


# 厳密解のプロット
x_analytical = np.linspace(0, b, 100)
y_analytical = np.exp(x_analytical * x_analytical * 0.5)
ax.plot(x_analytical,y_analytical,label='analytical', c="green")


# グラフタイトルなどの表示
ax.set_title('Euler')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(loc='best')
ax.grid(True) 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# データを定義します
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])
angles = np.array([0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi]) # 角度（ラジアン単位）

# 各点での矢印のX方向とY方向の成分を計算します
u = np.cos(angles)
v = np.sin(angles)

# 折れ線グラフを作成します
plt.plot(x, y, label='Trajectory')

# 各点に矢印を追加します
plt.quiver(x, y, u, v, scale=5, color='red')

# タイトルと軸ラベルを追加します
plt.title("Trajectory with Directional Arrows")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)

# グラフを表示します
plt.show()
```

```python
import numpy as np
class Perceptron(object):
    def __init__(self, x1, x2):
        self.x1 = x1
        self.x2 = x2

    def AND(self,and1,and2):
        x  = np.array([and1, and2])
        w1 = np.array([0.5, 0.5])
        b1 = -0.7
        tmp = np.sum(w1*x)+b1
        if tmp <= 0:
            print(0)
        else:
            print(1)
    def NAND(self):
        x  = np.array([self.x1, self.x2])
        w2 = np.array([-0.5, -0.5])
        b2 = 0.7
        tmp = np.sum(w2*x)+b2
        if tmp <= 0:
            print(0)
        else:
            print(1)
    def OR(self):
        x  = np.array([self.x1, self.x2])
        w1 = np.array([0.5, 0.5])
        b3 = -0.2
        tmp = np.sum(w1*x)+b3
        if tmp <= 0:
            print(0)
        else:
            print(1)

    def XOR(self):
       
        self.s1 = self.NAND()
        self.s2 = self.OR()
        self.y  = self.AND(self.s1, self.s2)
        if self.y == 1:
            print(1)
        else:
            print(0)
    

ito = Perceptron(0,1)
ito.AND(0,1)
ito.NAND()
ito.OR()
ito.XOR()
```

### 3プロット/フレーム

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# グラフで描画したいデータを指定
x = t  # 例として0から10までの100点
y = env.reward_graph  # 例として0から10のランダムな値

fig, ax = plt.subplots()
line, = ax.plot([], [], 'b')  # 初期線は空
# x軸とy軸の範囲をデータの最小値と最大値から計算し、余白を加える
x_margin = 0.05 * (max(x) - min(x))
y_margin = 0.05 * (max(y) - min(y))
ax.set_xlim(min(x) - x_margin, max(x) + x_margin)
ax.set_ylim(min(y) - y_margin, max(y) + y_margin)

ax.grid(True)  # グリッド線を追加

def init():
    line.set_data([], [])
    return line,

def update(frame):
    # frame を3倍して、3プロットずつ更新する
    end_index = min(frame * 10, len(x))  # インデックスがリストの長さを超えないように　＃変更箇所＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃
    line.set_data(x[:end_index], y[:end_index])
    return line,

# アニメーションの設定
# frames の数を調整して、全データが表示されるようにする
ani = FuncAnimation(fig, update, frames=(len(x) + 9) // 10, init_func=init, blit=True)# ＃変更箇所＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃

# 動画として保存
ani.save('plot_animation10.mp4', writer='ffmpeg', fps=30)

# 表示（Jupyter Notebook等で必要な場合）
plt.show()
```

### スキップフレーム指定可能

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# グラフで描画したいデータを指定
x = t  # 例として0から10までの100点
y = env.reward_graph  # 例として0から10のランダムな値
frameskip = 20

fig, ax = plt.subplots()
line, = ax.plot([], [], 'b')  # 初期線は空
ax.set_xlim(min(x), max(x))   # x軸の範囲設定
ax.set_ylim(min(y), max(y))   # y軸の範囲設定

def init():
    line.set_data([], [])
    return line,

def update(frame):
    # frame を3倍して、3プロットずつ更新する
    end_index = min(frame * frameskip, len(x))  # インデックスがリストの長さを超えないように　＃変更箇所＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃
    line.set_data(x[:end_index], y[:end_index])
    return line,

# アニメーションの設定
# frames の数を調整して、全データが表示されるようにする
ani = FuncAnimation(fig, update, frames=(len(x) + frameskip-1) // frameskip, init_func=init, blit=True)# ＃変更箇所＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃＃

# 動画として保存
ani.save('plot_animation10.mp4', writer='ffmpeg', fps=30)

# 表示（Jupyter Notebook等で必要な場合）
plt.show()
```
