```python
import matplotlib.pyplot as plt
import numpy as np

m = 1
b = 0.5
k = 1
shita = 0
initial_displacement = 1
initial_velocity = 0

# 関数の定義
def f(t):
    omega = np.sqrt(k/m - (b/(2*m))**2)
    return np.exp(-b/(2*m)*t) * (initial_displacement * np.cos(omega*t) + (initial_velocity + b/(2*m)*initial_displacement) * np.sin(omega*t))

# xの値の範囲を指定
x = np.linspace(0, 10, 100)

# グラフの描画
plt.plot(x, f(x))
plt.title('Damped Oscillation with Initial Conditions')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# 完成　振幅の理論式（解析解）
import matplotlib.pyplot as plt
import numpy as np

m = 1
damping = 0.5
k = 1
shita = 0
# 初期条件
x0 = 1  # 初期変位
v0 = 0  # 初速度

# 関数の定義
def f(t):
    omega = np.sqrt(k/m - (damping/(2*m))**2)
    return np.exp(-damping/(2*m)*t) * (x0 * np.cos(omega*t) + (v0 + damping/(2*m)*x0) * np.sin(omega*t))

# xの値の範囲を指定
x = np.linspace(0, 10, 1000)
y= f(x)
# グラフの描画
plt.plot(x, y)
plt.title("理論式による振幅の変化", fontname="MS Gothic")
plt.xlabel("時間t", fontname="MS Gothic")
plt.ylabel("振幅", fontname="MS Gothic")
plt.grid(True)
# plt.legend(prop={'family':'MS Gothic'})
plt.show()




# plt.grid()
# plt.title("z座標の変化", fontname="MS Gothic")
# plt.plot(times, cube1_zpos, linestyle='solid', marker='s', label="下のブロックのZ座標")
# plt.plot(times, cube2_zpos, linestyle='solid', marker='p', label="上のブロックのZ座標")
# plt.plot(times, cube2_greenball_zpos, linestyle='solid', marker='h', label="上のブロックについた緑玉のZ座標")
# # plt.plot(times, angular_velocity[2][2], linestyle='solid', marker='o')
# plt.legend(prop={'family':'MS Gothic'})
# plt.show()
```

```python
from PIL import Image, ImageDraw, ImageFont
import requests

# Load the image
url = "https://github.com/google-deepmind/mujoco/raw/main/banner.png"
response = requests.get(url)
img = Image.open(BytesIO(response.content))

# Create a drawing context
draw = ImageDraw.Draw(img)

# Define the text to be added
text = "ATSUSHI"

# Define the font family and size
font_family = "arial.ttf"
font_size = 36

# Define the font object
font = ImageFont.truetype(font_family, font_size)

# Define the width and height of the image
width, height = img.size

# Define the maximum font size
max_font_size = height // 4

# Define the minimum font size
min_font_size = 1

# Define the target width and height of the text
target_width = width * 0.8
target_height = height * 0.8

# Define the initial font size
font_size = max_font_size

# Define the font object
font = ImageFont.truetype(font_family, font_size)

# Define the size of the text
text_width, text_height = draw.textsize(text, font=font)

# Define the number of iterations
num_iterations = 100

# Define the tolerance
tolerance = 0.01

# Perform a binary search to find the optimal font size
for i in range(num_iterations):
    if abs(text_width - target_width) < tolerance and abs(text_height - target_height) < tolerance:
        break
    elif text_width > target_width or text_height > target_height:
        max_font_size = font_size
        font_size = (min_font_size + font_size) // 2
    else:
        min_font_size = font_size
        font_size = (max_font_size + font_size) // 2
    font = ImageFont.truetype(font_family, font_size)
    text_width, text_height = draw.textsize(text, font=font)

# Calculate the position of the text
x = (width - text_width) // 2
y = (height - text_height) // 2

# Add the text to the image
draw.text((x, y), text, font=font, fill=(255, 255, 255))

# Save the modified image
img.save("modified_bannerq.png")
```
